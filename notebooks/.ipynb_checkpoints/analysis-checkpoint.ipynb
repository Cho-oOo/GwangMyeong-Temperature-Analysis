```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the dataset
df = pd.read_csv("../data/gwangmyeong_temperature.csv")
```

```python
# Convert 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'])
```

```python
# Data cleaning: Remove any missing values
df = df.dropna()
```

```python
# Data exploration: Plotting the temperature trend
plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['Temperature'])
plt.title('Temperature Trend in Gwangmyeong (1970-2020)')
plt.xlabel('Year')
plt.ylabel('Temperature (°C)')
plt.savefig('../images/temperature_trend.png')
plt.show()
```
